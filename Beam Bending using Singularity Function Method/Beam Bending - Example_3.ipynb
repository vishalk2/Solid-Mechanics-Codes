```python
from sympy import *
from sympy.physics.continuum_mechanics.beam import Beam
```

```python
# Symbolizing Unknown quantities
E,I = symbols('E,I')
L,a = symbols('L,a')
w0 = symbols('w0')
RA,RB = symbols('RA,RB')
```

```python
# Defining beam
B = Beam(L,E,I)
```

```python
# Applying Loads
B.apply_load(RA,0,-1)
B.apply_load(-w0,0,0,end=a)
B.apply_load(RB,L,-1)
```

```python
B.load
```

```python
B.solve_for_reaction_loads(RA,RB)
B.reaction_loads
```

```python
B.load
```

```python
B.shear_force()
```

```python
B.bending_moment()
```
