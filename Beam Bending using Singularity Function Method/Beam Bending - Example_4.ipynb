```python
from sympy import *
from sympy.physics.continuum_mechanics.beam import Beam
```

```python
# Symbolizing unknown Quantities
E,I = symbols('E,I')
RA,RB = symbols('RA,RB')
```

```python
# Defining Beam
B = Beam(6,E,I)
```

```python
# Applying Loads
B.apply_load(RA,0,-1)
B.apply_load(12,2,-2)
B.apply_load(-12,4,-1)
B.apply_load(RB,6,-1)
```

```python
# Finding unknown loads
B.solve_for_reaction_loads(RA,RB)
B.reaction_loads
```

```python
B.load
```

```python
B.shear_force()
```

```python
B.bending_moment()
```

```python
# Plot of Shear Force variation VS x
B.plot_shear_force()
```

```python
# Plot of Bending Moment Variation VS x
B.plot_bending_moment()
```
